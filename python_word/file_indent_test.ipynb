```python
import os
import sys
import re
import datetime
from collections import Counter
```

```python
regex = re.compile(r'^[ ]{1,}')
```

```python
def traverse(folder_path):
    fs = os.listdir(folder_path)
    for fp in fs:
        tmp_path = os.path.join(folder_path, fp)
        flag = False
        if not(os.path.isdir(tmp_path)) and re.match(r'(.*)\.cs$', tmp_path) and 'obj' not in tmp_path:
            try:
                with open(tmp_path, 'r', encoding='utf-8') as fb:
                    for line in fb:
                        result = regex.search(line)
                        if result:
                            whitespace_len = len(result.group())
                            if whitespace_len >= 32 and not line.isspace():
                                print(line)
                                print(tmp_path)
                                break
            except:
                print("exception: ", tmp_path)
        elif os.path.isdir(tmp_path):
            traverse(tmp_path)
                    
```
