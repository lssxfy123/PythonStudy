双变量：散点图，热图，小提琴图

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
fuel_econ = pd.read_csv('fuel_econ.csv')
print(fuel_econ.shape)
fuel_econ.head()
```

```python
plt.scatter(data=fuel_econ, x='displ', y='comb')
plt.xlabel("Displacement(l)")
plt.ylabel("Combined Fuel Eff(spg)")
```

散点图：如果绘制的点太多，会发生重叠现象，如上图所示

```python
sb.regplot(data=fuel_econ, x='year', y='comb')
```

```python
## 对X轴进行抖动
## X轴数据随机增加[-0.3, 0.3]中的随机值
sb.regplot(data=fuel_econ, x='year', y='comb', x_jitter=0.3)
```

```python
## 修改透明度
sb.regplot(data=fuel_econ, x='year', y='comb', x_jitter=0.3, scatter_kws={'alpha': 1/20})
```

热图

```python
## 发动机排量与燃油效率
plt.hist2d(data=fuel_econ, x='displ', y='comb')
plt.colorbar()
plt.xlabel("Displacement(l)")
plt.ylabel("Combined Fuel Eff(spg)")
```

```python
plt.hist2d(data=fuel_econ, x='displ', y='comb', cmin=0.5)  # cmin指单元格达到什么样的最低值才会绘制，cmin=0.5表示至少有1个点才绘制
plt.colorbar()
plt.xlabel("Displacement(l)")
plt.ylabel("Combined Fuel Eff(spg)")
```

```python
## 对单元格的颜色进行翻转
plt.hist2d(data=fuel_econ, x='displ', y='comb', cmin=0.5, cmap='viridis_r')
plt.colorbar()
plt.xlabel("Displacement(l)")
plt.ylabel("Combined Fuel Eff(spg)")
```

```python
## 设置X轴，Y轴的组距
fuel_econ[['displ', 'comb']].describe()
```

```python
bins_x = np.arange(0.6, 7 + 0.3, 0.3)
bins_y = np.arange(12, 58 + 3, 3)
plt.hist2d(data=fuel_econ, x='displ', y='comb', cmin=0.5, cmap='viridis_r', bins=[bins_x, bins_y])
plt.colorbar()
plt.xlabel("Displacement(l)")
plt.ylabel("Combined Fuel Eff(spg)")
```

小提琴图

```python
fuel_econ.head()
```

```python
sedan_class = ['Minicompact Cars', 'Subcompact Cars', 'Compact Cars', 'Midsize Cars', 'Large Cars']
vclasses = pd.api.types.CategoricalDtype(ordered=True, categories=sedan_class)
print(vclasses)
```

```python
print(fuel_econ['VClass'].head())
fuel_econ['VClass'] = fuel_econ['VClass'].astype(vclasses)  # 强制类型转换，object转换为有序category
fuel_econ['VClass'].head()
```

```python
## 绘制小提琴图
base_color = sb.color_palette()[0]
sb.violinplot(data=fuel_econ, x='VClass', y='comb', color=base_color, inner=None)  # inner=None用来删除箱线图
plt.xticks(rotation=15)
```
